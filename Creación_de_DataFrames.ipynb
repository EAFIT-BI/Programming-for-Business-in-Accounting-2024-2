# Creación de *DataFrames*

Un *DataFrame* es el formato básico de información estructurada de `python` formado por *filas* y *columnas* con **encabezados** e **índices** opcionales y administrados por la librería `pandas`

```python
#importamos las librerías necesarias para la creación de la tabla

import pandas as pd

estudiantes = pd.DataFrame([['Andrea', 20, 'Mercadeo'],
 ['Kevin', 18, 'Contaduría'], ['Silvia', 19, 'Administración'],
['Pablo', 22, 'Administración'], ['Citlali', 20, 'Negocios'],
['Ana', 20, 'Contaduría'], ['Jhoan', 19, 'Contaduría'],['Jose', 16, 'Contaduría'],
['Nikolle', 18, 'Contaduría'],['Sandy', 17, 'Contaduría'], ['Andrea', 18, 'Contaduría'],
 ['Alexis', 20, 'Administración'], ['Juan Camilo', 20, 'Contaduría'],
 ['Tania', 17, 'Contaduría'],['Agua Marina', 20, 'Contaduría'],['Laura', 21, 'Negocios'],
 ['Salomé', 24, 'Contaduría'],['Carol', 20, 'Mercadeo'], ['Esteban', 19, 'Administración'],
 ['John', 17, 'Contaduría']])
```

```python

#lectura del archivo de excel con la información requerida
libro_diario =  pd.read_excel("/content/LibroDiario.xlsx")

#visualización de la información para efectos de comprobación
libro_diario.tail(6)
```

## Filtrado por columna

```python
#Filtramos la información para obtener únicamente la columna del débito
# la almacenamos en una columna llamada debito
debito = libro_diario['Débito']

#visualizamos para comprobar, en este caso haremos una impresión implícita
debito
```

```python
#sumamos los elementos del débito y los almacenamos en la variable
#suma débito
suma_debito = debito.sum()

#visualizamos para comprobar
suma_debito
```

```python
credito = libro_diario['Crédito']
suma_credito = credito.sum()
```

```python
# establecemos una condición para comparar el débito y el crédito

if suma_credito == suma_debito:
  print("El balance es correcto, la partida doble se cumple")
else:
  print("Póngase mosca!")
```

## Filtrados por contenido

Supongamos que la junta directiva está interesada en conocer los Activos Corrientes de la empresa en un periodo

```python
#primero realizamos el filtrado
activos_corrientes = libro_diario[libro_diario['A - PS - PT'] == "AC"]

#visualizamos para comprobar
activos_corrientes
```

```python
#podemos realizar un segundo filtrado por columna para obtener información
#más relevante. Si voy a crear una nueva tabla con menos columnas le pongo
#doble corchete
datos_relevantes = activos_corrientes[['Fecha', 'Nombre Cta', 'Débito', 'Crédito']]

#visualizamos
datos_relevantes
```
