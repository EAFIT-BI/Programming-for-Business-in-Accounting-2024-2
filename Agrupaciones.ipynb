# Agrupaciones

Las agrupaciones sirven para contraer la información de una tabla. Siempre acompañadas de una medida de resumen. Dado que el resultado es una tabla, es posible graficarlo y volver a resumirlo

```python
#importamos librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cargamos la información
empresas = pd.read_excel('/content/Base-1000-empresas-2024.xlsx',
                         header = 4)
```

```python
#visualizamos la información
empresas.head()
#imprimos las columnas
empresas.columns
```

```python
empresas[['SUPERVISOR', 'INGRESOS OPERACIONALES\n2023*']].groupby('SUPERVISOR').mean()
```

```python
empresas[['REGIÓN', 'MACROSECTOR']].groupby('REGIÓN').agg(pd.Series.mode)
```

```python
resultados = empresas[['REGIÓN', 'SUPERVISOR']].groupby('REGIÓN').count()
```

```python
#graficamos resultados
resultados.plot.bar();
```

```python
#aplicamos una doble agregación o medida de resumen.
#primero con la agrupación y luego con el filtrado
resultados[resultados['SUPERVISOR'] == resultados['SUPERVISOR'].max()]
```

```python
contadores[['Tipo', 'Proceso Jurídico']].groupby('Tipo').count()
```
